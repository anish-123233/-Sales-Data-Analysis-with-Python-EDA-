#### Counting pairs of products that were sold together(for recommendations or make smart deals)


```python
df[df['Grouped'].apply(lambda x: len(x.split(','))==2)]['Grouped'].value_counts().head(20)
```

```python
from itertools import combinations  # Used to generate all possible pairs from a list
from collections import Counter  # Used to count occurrences of elements

# Initialize a Counter object to store the frequency of product pair occurrences
count = Counter()

# Iterate over each row in the 'Grouped' column of the DataFrame
for row in df['Grouped']:
    # Split the comma-separated product names into a list
    row_list = row.split(',')
    
    # Generate all possible pairs (combinations of 2 products) from the list
    product_combinations = combinations(row_list, 2)
    
    # Update the Counter with the generated product pairs
    count.update(Counter(product_combinations))

# Print the 10 most common product pairs
count.most_common(10)
```
